```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import difflib 
```

```python
#read in the data
data_tests = pd.read_csv('data-private-school-tests.csv')
data_tests 
```

```python
data_matches = pd.read_csv('private_successful_matches.csv')
data_matches
```

```python
print("Data from Tests file:")
print(data_tests.head())
```

```python
print("\nData from Matches file:")
print(data_matches.head())
```

```python
data_tests['School'] = data_tests['School'].str.lower()
```

```python
data_matches['Matched School Name'] = data_matches['Matched School Name'].str.lower()
```

```python
matched_schools_adjusted = pd.merge(data_tests, data_matches, left_on='School', right_on='Matched School Name', how='inner')
```

```python
matched_schools_adjusted_info = matched_schools_adjusted.head()
```

```python
matched_schools_adjusted_count = matched_schools_adjusted.shape[0]
```

```python
matched_schools_adjusted_info, matched_schools_adjusted_count
```

```python
unmatched_schools = data_tests[~data_tests['School'].isin(matched_schools_adjusted['School'])]
```

```python
# Display the unmatched schools
unmatched_schools_info = unmatched_schools.drop_duplicates(subset=['School'])  # Dropping duplicate entries for clarity
unmatched_schools_info
```

```python
matched_schools_adjusted.to_csv('matched_schools.csv', index=False)
```

```python
unmatched_schools_info.to_csv('unmatched_schools.csv', index=False)
```
