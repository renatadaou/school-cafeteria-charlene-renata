```python
# prepare to read csv and use pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv "Private School Complete Merged Data.csv"
df = pd.read_csv("Private School Complete Merged Data.csv")
df
```

```python
# in column "PSS_LIBRARY", replace "Yes" with "1" and "No" with "0"
df['PSS_LIBRARY'] = df['PSS_LIBRARY'].replace('Yes', 1)
df['PSS_LIBRARY'] = df['PSS_LIBRARY'].replace('No', 0)
df
```

```python
# create a column "Jewish" with 1 if "PSS_ORIENT" = 18, 0 otherwise
df['Jewish'] = np.where(df['PSS_ORIENT'] == 18, 1, 0)
df
```

```python
# create a column "Catholic" with 1 if "PSS_RELIG" = 1, 0 otherwise
df['Catholic'] = np.where(df['PSS_RELIG'] == 1, 1, 0)
df
```

```python
# create a column "Religious" with 0 if "PSS_RELIG" = 3, 1 otherwise
df['Religious'] = np.where(df['PSS_RELIG'] == 3, 0, 1)
df
```

```python
# save to csv "Private School Complete Merged Data - cleaned.csv"
df.to_csv("Private School Complete Merged Data - cleaned.csv", index=False)
```

```python
%%R
# read "Private School Complete Merged Data - cleaned.csv" into R
df <- read_csv("Private School Complete Merged Data - cleaned.csv")
df
```

```python
%%R
# remove "%" in "Proficiency" and convert to numeric
df$Proficiency <- as.numeric(gsub("%", "", df$Proficiency))
df
```

```python
%%R
# conduct a simple anaylsis of the data
summary(df)
```

```python
%%R
# show each column name
names(df)
```

```python
%%R
# plot a histogram of "Proficiency"
df %>%
  ggplot(aes(Proficiency)) +
  geom_histogram(bins = 30, fill = "grey", color = "black") +
  labs(title = "Histogram of Proficiency", x = "Proficiency", y = "Frequency")
```

```python
%%R
# show all the data for "Violation counts"
df %>%
  select("Violation counts")
```

```python
%%R
# plot the distribution of "Violation counts"
df %>%
  ggplot(aes(`Violation counts`)) +
  geom_histogram(bins = 30, fill = "grey", color = "black") +
  labs(title = "Histogram of Violation counts", x = "Violation counts", y = "Frequency")
```

```python
%%R
# filter to "Jewish == 1" and show the data
df %>%
  filter(Jewish == 1)
```

```python
%%R
# plot the distribution of "Proficiency" for "Jewish == 1"
df %>%
  filter(Jewish == 1) %>%
  ggplot(aes(Proficiency)) +
  geom_histogram(bins = 30, fill = "grey", color = "black") +
  labs(title = "Histogram of Proficiency for Jewish == 1", x = "Proficiency", y = "Frequency")
```

```python
%%R
# plot the distribution of "Violation counts" for "Jewish == 1"
df %>%
  filter(Jewish == 1) %>%
  ggplot(aes(`Violation counts`)) +
  geom_histogram(bins = 30, fill = "grey", color = "black") +
  labs(title = "Histogram of Violation counts for Jewish == 1", x = "Violation counts", y = "Frequency")
```

```python
%%R
# plot the density of "Violation counts" for "Jewish == 1" and "Jewish == 0"
df %>%
  ggplot(aes(`Violation counts`, color = factor(Jewish))) +
  geom_density() +
  labs(title = "Density of Violation counts for Jewish == 1 and Jewish == 0", x = "Violation counts", y = "Density")
  
```

```python
%%R
# conduct a multiple regression of "Violation counts" on "Jewish", and "Proficiency"
# corrected to avoid the parser error
fit <- lm(`Violation counts` ~ Jewish + Proficiency, data = df)
summary(fit)
```
